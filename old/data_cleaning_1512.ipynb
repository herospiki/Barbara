```python
MAPPING_CORRECTIF =  {
    'acalmie':'accalmie', 
     'bruine puis très beau': 'bruine/très beau', 
     'beau puis bruine' : 'beau/bruine', 
     'averses, éclaircies' : 'averses/éclaircies',
    'soleil et qq nuages' : 'soleil/qq nuages', 
    'bruine le matin': 'bruine/mitigé',
    'fortes pluie' : 'fortes pluies',
    'éclairicie':'éclaircies'
}

def corriger_meteo(terme):
    if pd.isna(terme) or terme == '' or terme is None:
        return ''
    
    # Nettoyage de base : minuscule, retrait des espaces
    terme_nettoye = terme.lower().strip()
    
    # Essayer de mapper le terme nettoyé
    return MAPPING_CORRECTIF.get(terme_nettoye,terme)

df['Météo_Corrigée'] = df['Météo'].apply(corriger_meteo)

df = df.drop(columns=['Météo']).rename(columns={'Météo_Corrigée': 'Météo'})

```

### Fonctions de traitement des pontes

```python
def traiter_une_ponte(val):
    """
    Traite une valeur unique de Ponte.
    Retourne un dictionnaire avec: Ponte, Etat_oeuf, Doute, Statut_Poule, Remarques
    """
    
    # Valeurs par défaut
    resultat = {
        'Ponte': 0,
        'Etat_oeuf': 'RAS',
        'Doute': False,
        'Statut_Poule': 'Présente',
        'Remarques': ''
    }
    
    # Convertir en string et nettoyer les chevrons invisibles
    val = str(val).strip().replace('\xa0', '')
    
    # Cas NaN ou vide
    if val == 'nan' or val == '':
        return resultat
    
    # Cas 1: Valeurs numériques avec annotations n(+m)
    if re.match(r'^\d+\(\+\d+\)$', val):
        match = re.match(r'^(\d+)\(\+(\d+)\)$', val)
        n = int(match.group(1))
        m = int(match.group(2))
        resultat['Ponte'] = n + m
        resultat['Remarques'] = f"{m} oeufs trouvés dehors"
        return resultat
    
    # Cas 2: Valeurs avec 'x' et possibles annotations
    if 'x' in val.lower():
        # Cas: pur nombre suivi de 'x' (ex: '2x')
        if re.match(r'^\d+x$', val.lower()):
            nombre = int(val[:-1])
            resultat['Ponte'] = nombre
            return resultat
        
        # Cas: 'x' avec parenthèses (ex: 'x(mou)', 'x(cassé)')
        if re.match(r'^x\(.*\)$', val.lower(), re.IGNORECASE):
            resultat['Ponte'] = 1
            etat = re.search(r'\((.*?)\)', val)
            if etat:
                resultat['Etat_oeuf'] = etat.group(1)
            return resultat
        
        # Cas: 'x?' avec possibles parenthèses (ex: 'x?(mou)', 'x?(cassé)')
        if re.match(r'^x\?(?:\(.*?\))?$', val.lower(), re.IGNORECASE):
            resultat['Ponte'] = 1
            resultat['Doute'] = True
            etat = re.search(r'\((.*?)\)', val)
            if etat:
                resultat['Etat_oeuf'] = etat.group(1)
            return resultat
        
        # Cas: pur 'x' ou 'x '
        if re.match(r'^x\s*$', val.lower()):
            resultat['Ponte'] = 1
            return resultat
    
    # Cas: '?' seul
    if val == '?':
        resultat['Doute'] = True
        return resultat
    
    # Cas spécial: DCD (Décédée)
    if val.upper() == 'DCD':
        resultat['Statut_Poule'] = 'Décédée'
        return resultat
    
    # Cas 3: Valeurs numériques pures (0, 1, 2, 3)
    if re.match(r'^\d+$', val):
        resultat['Ponte'] = int(val)
        return resultat
    
   
    # Cas 4: Mots seuls (tous les autres)
    resultat['Remarques'] = val
    return resultat
```

```python
def traiter_dataframe_ponte(df, colonnes_poules):
    """
    Traite un DataFrame avec plusieurs colonnes de poules.
    
    Paramètres:
    - df: DataFrame original
    - colonnes_poules: liste des noms de colonnes de poules
    
    Crée pour chaque poule 5 colonnes:
    - {poule}_Ponte
    - {poule}_Etat_oeuf
    - {poule}_Doute
    - {poule}_Statut_Poule
    - {poule}_Remarques
    """
    
    df_result = df[['Date']].copy()  # Garder la colonne Date
    
    for poule in colonnes_poules:
        # Appliquer le traitement à chaque valeur de la colonne
        donnees_traitees = df[poule].apply(traiter_une_ponte)
        
        # Extraire chaque champ dans une colonne séparée
        df_result[f'{poule}_Ponte'] = donnees_traitees.apply(lambda x: x['Ponte'])
        df_result[f'{poule}_Etat_oeuf'] = donnees_traitees.apply(lambda x: x['Etat_oeuf'])
        df_result[f'{poule}_Doute'] = donnees_traitees.apply(lambda x: x['Doute'])
        df_result[f'{poule}_Statut_Poule'] = donnees_traitees.apply(lambda x: x['Statut_Poule'])
        df_result[f'{poule}_Remarques'] = donnees_traitees.apply(lambda x: x['Remarques'])
    
    return df_result
```

```python
def propager_statut_deces(series_statut):
    """
    Propager le statut 'Décédée' pour toutes les lignes
    après la première occurrence dans cette série.
    
    Exemple:
    ['Présente', 'Présente', 'Décédée', 'Présente'] 
    -> ['Présente', 'Présente', 'Décédée', 'Décédée']
    """
    
    a_deceder = series_statut == 'Décédée'
    
    if a_deceder.any():
        # Trouver la POSITION (index entier) de la première occurrence
        premier_deces_position = a_deceder.argmax()
        
        # Mettre à jour toutes les valeurs à partir de cette position
        series_statut.iloc[premier_deces_position:] = 'Décédée'
    
    return series_statut
```

```python
def completer_donnees_ponte_wide(df_wide, colonnes_poules):
    """
    Complète les données de ponte au format wide (avant conversion en long).
    
    Étapes:
    1. Quand une poule 2024 n'est pas présente en 2023, on considère qu'elle est 'Absente' en 2023 et on
    met son statut à 'Absente' pour toutes les dates de 2023.
    2. Propager le statut 'Décédée' pour chaque poule
    3. Forcer Ponte=0 quand Statut_Poule='Décédée'

    
    Paramètres:
    - df_wide: DataFrame avec colonnes {poule}_Ponte, {poule}_Statut_Poule, etc.
    - colonnes_poules: liste des noms de poules
    
    Retourne: DataFrame complété
    """
    
    df_wide = df_wide.copy()
    
    for poule in colonnes_poules:
        col_statut = f'{poule}_Statut_Poule'
        col_ponte = f'{poule}_Ponte'
        
        # Propager le statut 'Décédée'
        df_wide[col_statut] = propager_statut_deces(df_wide[col_statut])
        
        # Forcer Ponte=0 quand Décédée
        df_wide.loc[df_wide[col_statut] == 'Décédée', col_ponte] = 0

    
    
    print(f"✅ Données complétées (format wide): {len(df_wide)} lignes")
    print(f"   Poules traitées: {', '.join(colonnes_poules)}")
    
    return df_wide
```

```python
def convertir_format_long(df_wide, colonnes_poules):
    """
    Convertit le DataFrame du format wide au format long (pour Tableau).
    
    Paramètres:
    - df_wide: DataFrame avec colonnes {poule}_Ponte, {poule}_Etat_oeuf, etc.
    - colonnes_poules: liste des noms de poules
    
    Retourne un DataFrame avec colonnes: Date, Poule, Ponte, Etat_oeuf, Doute, Statut_Poule, Remarques
    """
    
    data_long = []
    
    for _, row in df_wide.iterrows():
        date = row['Date']
        
        for poule in colonnes_poules:
            data_long.append({
                'Date': date,
                'Poule': poule,
                'Ponte': row[f'{poule}_Ponte'],
                'Etat_oeuf': row[f'{poule}_Etat_oeuf'],
                'Doute': row[f'{poule}_Doute'],
                'Statut_Poule': row[f'{poule}_Statut_Poule'],
                'Remarques': row[f'{poule}_Remarques']
            })
    
    df_long = pd.DataFrame(data_long)
    
    # Réorganiser les colonnes dans un ordre logique
    df_long = df_long[['Date', 'Poule', 'Ponte', 'Etat_oeuf', 'Doute', 'Statut_Poule', 'Remarques']]
    
    return df_long
```

```python
colonnes_poules = ['Joséphine', 'Augustine', 'Cunégonde', 'Pioupioute', 'Valérie',
       'Rémiel', 'Saquiel', 'Albertine', 'Nina et Tina', 'Marans', 'Nina','Tina']

df_traite = traiter_dataframe_ponte(df, colonnes_poules)

# 2. Compléter le statut
df_complete = completer_donnees_ponte_wide(df_traite, colonnes_poules)
for poule in colonnes_poules:
    if poule not in poules_2023:
        # Poule absente en 2023
        df_complete.loc[df_complete['Date'] < '2024-01-01', f'{poule}_Statut_Poule'] = 'Absente'
        # Forcer Ponte=0 quand Absente
        df_complete.loc[df_complete[f'{poule}_Statut_Poule'] == 'Absente', f'{poule}_Ponte'] = 0
        
# 3. Convertir en format long pour Tableau
df_long = convertir_format_long(df_complete, colonnes_poules)
df_long.head()

# Exporter pour Tableau
#df_long.to_csv('ponte_data_tableau.csv', index=False)
```

```python

def attribuer_nina(df, seed=42):
    """
    Attribue les pontes à Nina avec les contraintes :
    1. Chaque poule pond max 1 œuf/jour
    2. Total par jour (Albertine + Tina + Nina) <= 3
    3. Albertine est certaine
    4. Tina : certaine si Tina=1, aléatoire (50/50) si "Nina et Tina"=1 et Tina=0
    5. Les excédents Marans remontent aux jours précédents SANS VIOLER la contrainte
    
    Paramètres:
    - df: dataframe avec colonnes ['Date', 'Albertine', 'Marans', 'Nina et Tina', 'Nina','Tina']
    - seed: pour reproductibilité
    """
    
    if seed is not None:
        np.random.seed(seed)
    
    df = df.copy()
    n = len(df)
    
    # Étape 1 : Gérer l'attribution aléatoire pour "Nina et Tina"
    df['Tina_attr'] = df['Tina'].copy()
    df['Nina'] = 0
    
    # Où Tina = 0 et Nina et Tina = 1, c'est une attribution aléatoire
    mask_aleatory = (df['Tina'] == 0) & (df['Nina et Tina'] == 1)
    df.loc[mask_aleatory, 'Tina_attr'] = np.random.randint(0, 2, size=mask_aleatory.sum())
    df.loc[mask_aleatory, 'Nina'] = 1 - df.loc[mask_aleatory, 'Tina_attr']
    
    # Étape 2 : Propager les excédents Marans vers les jours précédents
    marans_attr = df['Marans'].values.copy().astype(float)
    nina_attr = df['Nina'].values.copy().astype(float)
    tina_attr = df['Tina_attr'].values.copy().astype(float)
    
    # Remonter de la fin vers le début
    for i in range(n - 1, 0, -1):
        albertine = df.iloc[i]['Albertine']
        tina = tina_attr[i]
        
        # Place disponible pour Nina aujourd'hui
        disponible_nina = max(0, 3 - albertine - tina)
        
        # Oeufs Marans à attribuer
        oeufs_marans = marans_attr[i]
        
        # Attribution à Nina (max 1)
        nina_attr[i] = min(1, min(disponible_nina, max(0, oeufs_marans - albertine - tina)))
        
        # Excédent à remonter
        excedent = oeufs_marans - albertine - tina - nina_attr[i]
        
        if excedent > 0:
            # Vérifier que le jour précédent peut l'accepter
            albertine_prev = df.iloc[i-1]['Albertine']
            tina_prev = tina_attr[i-1]
            nina_prev = nina_attr[i-1]
            
            place_dispo_prev = max(0, 3 - albertine_prev - tina_prev - nina_prev)
            
            # Remonter ce qu'on peut
            excedent_remonte = min(excedent, place_dispo_prev)
            marans_attr[i-1] += excedent_remonte
            
            # Ce qui n'a pas pu remonter... reste aujourd'hui (problème non résolu)
            # On le signale mais on l'attribue quand même à Nina si possible
            if excedent > excedent_remonte:
                # On essaie d'attribuer plus à Nina aujourd'hui
                supplementaire = min(1, excedent - excedent_remonte)
                if nina_attr[i] + supplementaire <= 1:
                    nina_attr[i] += supplementaire
    
    # Traiter le premier jour
    albertine_0 = df.iloc[0]['Albertine']
    tina_0 = tina_attr[0]
    disponible_nina_0 = max(0, 3 - albertine_0 - tina_0)
    nina_attr[0] = min(1, min(disponible_nina_0, max(0, marans_attr[0] - albertine_0 - tina_0)))
    
    df['Attribution_Nina'] = nina_attr.astype(int)
    
    # Vérifier les contraintes
    df['Total_jour'] = df['Albertine'] + df['Tina_attr'] + df['Attribution_Nina']
    
    return df

# Exemple d'utilisation :
# df = pd.read_csv('vos_donnees.csv')
# df = attribuer_nina(df, seed=42)
# print(df[['Date', 'Albertine', 'Marans', 'Tina_attr', 'Attribution_Nina', 'Total_jour']])
```
